```python
import numpy as np

def f(x):
    return (x[0]**2 + x[1]**2 - 50)**2 + (x[0] - x[1])**2

def cuckoo_search(n=25, pa=0.25, iterations=60):
    L = np.random.uniform(-10, 10, (n, 2))
    fitness = np.array([f(x) for x in L])
    best = L[np.argmin(fitness)]

    for _ in range(iterations):
        for i in range(n):
            step = np.random.normal(0, 1, 2)
            levy = 0.01 * step / (np.abs(step)**1.5)
            new = L[i] + levy
            if f(new) < fitness[i]:
                L[i], fitness[i] = new, f(new)

        for i in range(n):
            if np.random.rand() < pa:
                L[i] = np.random.uniform(-10, 10, 2)
                fitness[i] = f(L[i])

        if f(best) > np.min(fitness):
            best = L[np.argmin(fitness)]

    return best, f(best)

best, val = cuckoo_search()
best = np.round(best, 3)

print("Best solution:", list(best))
print("Objective function value:", float(val))
```
