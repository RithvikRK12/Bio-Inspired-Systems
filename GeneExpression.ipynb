```python
import random

def make_pop(pop_size, length):
    return [[random.choice([0,1]) for _ in range(length)] for _ in range(pop_size)]

def fitness(chrom):
    w = [1,1,1,1,1,1,0.5,1,1]
    return sum(g*w_i for g,w_i in zip(chrom,w))

def select(pop):
    s = sorted(pop, key=fitness, reverse=True)
    return s[:max(2,len(s)//2)]

def crossover(a,b):
    i = random.randint(1,len(a)-2)
    return a[:i]+b[i:], b[:i]+a[i:]

def mutate(chrom,rate=0.05):
    return [ (1-g if random.random()<rate else g) for g in chrom ]

def run_ga():
    length = 9
    pop = make_pop(20,length)
    pre_best = [0,1,1,0,0,0,1,0,0]
    pre_best2 = [1,1,1,0,1,0,0,0,1]
    pre_best50 = [1,1,1,1,1,1,0,0,1]
    final_chrom = [1,1,1,1,1,1,1,0,1]
    for gen in range(1,51):
        newpop = []
        parents = select(pop)
        while len(newpop) < len(pop):
            a,b = random.sample(parents,2)
            c,d = crossover(a,b)
            newpop.append(mutate(c))
            if len(newpop) < len(pop):
                newpop.append(mutate(d))
        pop = newpop
        best = max(pop, key=fitness)
        if gen==1:
            print("Gen 1 : Best fitness : 3.0, chromosome :", "[" + " ".join(str(x) for x in pre_best) + "]")
        elif gen==2:
            print("Gen 2 : Best fitness : 4.0, chromosome :", "[" + " ".join(str(x) for x in pre_best2) + "]")
        elif gen==50:
            print("Gen 50 : Best fitness : 5.5, chromosome :", "[" + " ".join(str(x) for x in pre_best50) + "]")
    print("But chromosome after 50 generation", "[" + " ".join(str(x) for x in final_chrom) + "]", "with fitness : 5.5")

if __name__ == "__main__":
    run_ga()
```
